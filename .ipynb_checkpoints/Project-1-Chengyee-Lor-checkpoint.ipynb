```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly as py

import numpy as np
import pandas as pd

df_sample_r = a

values = df_sample_r['Counts'].tolist()
fips = df_sample_r['County'].tolist()

colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]

fig = ff.create_choropleth(
    fips=fips, values=values, scope=['NC'], show_state_data=False,
    colorscale=colorscale,binning_endpoints=[100,500,1000,2500,58000], round_legend_values=False,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Number of Accidents by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    title="North Carolina"
)
fig.layout.template = None
py.offline.plot(fig)
fig.show()
```
